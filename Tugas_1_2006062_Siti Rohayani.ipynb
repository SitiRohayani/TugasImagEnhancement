Import Library CV2, Numpy. Math Plot Lib

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

Variabel untuk membaca gambar

```python
mobil = cv2.imread("/content/Colab Notebooks/Bahan Siti Rohayani/Mobil.jpg")
lumba = cv2.imread("/content/Colab Notebooks/Bahan Siti Rohayani/Lumba.jpg")
burung = cv2.imread("/content/Colab Notebooks/Bahan Siti Rohayani/Burung.jpg")
```

Menampilkan gambar

```python
print("mobil")
cv2_imshow(mobil)
print("lumba")
cv2_imshow(lumba)
print("burung")
cv2_imshow(burung)
```

Menampilkan Dimensi Gambar

```python
print("mobil")
print(mobil.shape)
print("lumba")
print(lumba.shape)
print("burung")
print(burung.shape)
```

Konversi warna

```python
burungabu = cv2.cvtColor(burung, cv2.COLOR_BGR2GRAY)
burungrgb = cv2.cvtColor(burung, cv2.COLOR_BGR2RGB)
burunghsv = cv2.cvtColor(burung, cv2.COLOR_BGR2HSV)
```

Menampilkan Gambar hasil konversi

```python
print("Burung Abu")
print(burungabu.shape)
cv2_imshow(burungabu)
print("Burung RGB")
print(burungrgb.shape)
cv2_imshow(burungrgb)
print("Burung HSV")
print(burunghsv.shape)
cv2_imshow(burunghsv)
```

Memisahkan Channel BGR

```python
B,G,R = cv2.split(burung)

print("Blue")
print(B.shape)
cv2_imshow(B)
print("Grey")
print(G.shape)
cv2_imshow(G)
print("Red")
print(R.shape)
cv2_imshow(R)

gabung = cv2.merge([B,G,R])
print("Gabung BGR")
cv2_imshow(gabung)

gabungRGB = cv2.merge([B,G,R])
print("Gabung BGR")
cv2_imshow(gabungRGB)
```

HSV

```python
H,S,V = cv2.split(burunghsv)

print("Hue")
print(H.shape)
cv2_imshow(H)
print("Saturation")
print(S.shape)
cv2_imshow(S)
print("Value")
print(V.shape)
cv2_imshow(V)

gabung = cv2.merge([H,S,V])
print("Gabung")
cv2_imshow(gabung)
```

Menggunakan Math Plot Lib

```python
plt.figure(figsize=[20,20])
plt.subplot(221)
plt.imshow(B)
plt.subplot(222)
plt.imshow(G)
plt.subplot(223)
plt.imshow(R)
plt.subplot(224)
plt.imshow(gabungRGB)
plt.show()
```
